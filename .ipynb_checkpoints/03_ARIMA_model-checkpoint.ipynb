# Libraries

```python
import pandas as pd
import numpy as np
import math as mt
from sklearn.metrics import mean_squared_error
import pmdarima as pm
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
sns.set()
```

# Metrics and Baseline

We will keep using the same metrics in all models. just remembering why we chose our metrics:

A percentage error makes no sense when measuring the accuracy of temperature forecasts on Celsius scales, because temperature has an arbitrary zero point. Therefore, we will use the average error (bias) to look at the overall direction of the forecast, and to mesure the precision of the forecast, we will use the mean absolute error (MAE) and root mean squared error (RMSE).

```python
#func to return a df with all our metrics

def metrics(serie_true,serie_pred,model_name ='Model'):
    err = [v_true - v_false for v_true,v_false in zip(serie_true,serie_pred)]
    absolute_err = [abs(i) for i in err]
    
    bias = np.mean(err)
    mae = np.mean(absolute_err) 
    rmse = mt.sqrt(mean_squared_error(serie_true, serie_pred)) 
    
    metrics_df = pd.DataFrame({'Bias':bias,'MAE':mae,'RMSE':rmse},index =[model_name])
    metrics_df.index.name = 'Model'
    return metrics_df
```

For the Baseline, now we have not only the naïve forecas but also, the Exponential Smoothing seasonal model.

Please note that development of Exponential Smoothing and naïve models, as well as reasons for choosing each, were already presented in file 02. So here we will just import the resulting metrics to use as baseline and comparison.

```python
baseline_metrics = pd.read_csv('Metrics/exp_smoothing_metrics.csv')
baseline_metrics.set_index('Model',inplace = True)

baseline_metrics
```

# ARIMA Model

```python
# loadin our pre-treated df 

min_temp_df = pd.read_csv('Datasets/daily-minimum-temperatures-in-me-treated_v2.csv')
display(min_temp_df.head(3))
min_temp_df.info()
```

```python
#date to datetime type and index

min_temp_df.date = pd.to_datetime(min_temp_df.date, format="%Y/%m/%d") 
min_temp_df.set_index('date',inplace = True)
```

We know that our time series has a strong sesanality, thus, it is not stationary.

However, we have a daily forecast, and an annual seasonality, which means that we have a seasonal period of 365 records, and a seasonal model may not handle such a large period well (SARIMA for example has a maximum value of 350 for a seasonal period) besides, as explained in the Exponential Smoothing model, it would not make sense to say that the forecast of the lowest daily temperature depends on the value that occurred on the exact same day of the previous year.

Therefore, we will try to apply the classic ARIMA model (which is usually applied only to stationary series) the justification is that for daily records our series can be interpreted as close to a stationary series. To test this claim we will apply the Dickey-Fuller test.

```python
# Dickey-Fuller test

result = adfuller(min_temp_df.daily_min_temp)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

#if the p-value is lower then 5% (0.05) it indicates that the serie is stationary
```

It seens like we were right, we got an p-value << 5%. We know that even with this low value, our serie isn't stationary. However, the data does not have a unit root ,so we will try to treat it as stationary.

```python
#Using auto-arima to choose the best parameters

model = pm.auto_arima(min_temp_df.daily_min_temp, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=True,   # Seasonality 
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)

print(model.summary())
```

```python
model.plot_diagnostics(figsize=(8,6))
plt.tight_layout()
plt.show()
```

Top left: The residual errors seem to fluctuate around a mean of zero and have a uniform variance.

Top Right: The density plot suggest normal distribution with mean zero.

Bottom left: All the dots should fall perfectly in line with the red line. Any significant deviations would imply the distribution is skewed.

Bottom Right: The Correlogram, aka, ACF plot shows the residual errors are not autocorrelated. Any autocorrelation would imply that there is some pattern in the residual errors which are not explained in the model. So you will need to look for more X’s (predictors) to the model.

Looks like we have a good model!

To evaluate the ARIMA model we will forecast day min temperature for each record (in exception for the first year)

To do this correctly, we will train the model only with the data collected before the forecast date. That is, we will train a model for each forecast, dropping the previous data.

```python

def evaluate_arima_forecast(df,col_forecast = 'daily_min_temp', p =3 ,d=0 ,q=1):
    counter = 0
    date_forecast, forecast,true_value = [[] for i in range(3)]
     
    for n in range(1,len(df)-365):
        train_df = df.drop(df.tail(n).index) #drop n last values to fit the model with right data
        train_df.reset_index(drop= True, inplace = True)
        model = sm.tsa.arima.ARIMA(train_df[col_forecast], order=(p, d, q))  
        fitted = model.fit(method_kwargs={"warn_convergence": False}) 
        forecast_result = fitted.predict(len(train_df))
        
        forecast.append(forecast_result.values[0])
        true_value.append(df[col_forecast][len(df)-n])
        date_forecast.append(df.index[len(df)-n])
        
        #Loading status
        percentage = mt.ceil((100*n)/(len(df)-365))
        if percentage%10 == 0:
            if counter!=percentage:
                print(f'{percentage}% completed')
                counter = percentage
        
    forecast_df = pd.DataFrame({col_forecast:true_value,'forecast':forecast},index =date_forecast)
    forecast_df= forecast_df.iloc[::-1] #reverse df to keep dates in chronological order
    return forecast_df
```

```python
arima_forecast_df = evaluate_arima_forecast(min_temp_df)
arima_forecast_df.head()
```

```python
arima_forecast_df.to_csv('arima_forecast.csv')
```

```python
#preparing formatting for data viz

def celsius_formatter(x, pos):
    'The two args are the value and tick position'
    return '%1.1f°C' % (x)

celsius_formatter = FuncFormatter(celsius_formatter)

plt.rcParams.update({'ytick.labelsize': 16.0,
                     'xtick.labelsize': 16.0,
                     'axes.labelsize': 16.0,
                     'axes.xmargin': 0.0})

sns.set_style("white")
```

```python
#Vizualizing the forecast 

fig, (ax1,ax2) = plt.subplots(2,1,figsize=(16,16))

sns.lineplot(x=arima_forecast_df.index,
             y=arima_forecast_df.daily_min_temp,
            color ='#32a852',lw =0.5,ax = ax1)

sns.lineplot(x=arima_forecast_df.index,
             y=arima_forecast_df.forecast,
            color ='#cf1919',lw =0.5,ax = ax1)

ax1.set_title("Daily minimum temperatures in Melbourne (Real x Forecast) 1982-1990",fontsize = 24)
ax1.set_xlabel('Year',fontsize = 22) #caracteristica das labels 
ax1.set_ylabel('Temperature',fontsize = 22)
ax1.yaxis.set_major_formatter(FuncFormatter(celsius_formatter))
ax1.legend(['Real','Forecast'],fontsize =16)

#################fig 2

arima_forecast_1990 = arima_forecast_df[arima_forecast_df.index.year ==1990]

sns.lineplot(x=arima_forecast_1990.index,
             y=arima_forecast_1990.daily_min_temp,
            color ='#32a852',lw =0.5,ax = ax2)

sns.lineplot(x=arima_forecast_1990.index,
             y=arima_forecast_1990.forecast,
            color ='#cf1919',lw =0.5,ax = ax2)

ax2.set_title("Daily minimum temperatures in Melbourne (Real x Forecast) 1990",fontsize = 24)
ax2.set_xlabel('',fontsize = 22) #caracteristica das labels 
ax2.set_ylabel('Temperature',fontsize = 22)
ax2.yaxis.set_major_formatter(FuncFormatter(celsius_formatter))
ax2.legend(['Real','Forecast'],fontsize =16)

plt.show()
```

```python
#Model Metrics

arima_metrics =  metrics(arima_forecast_df['daily_min_temp'],arima_forecast_df['forecast'],model_name ='ARIMA (3,0,1)')
arima_metrics
```

```python
# Combining the results with our baseline

metrics_df = pd.concat([baseline_metrics,arima_metrics])
metrics_df.to_csv('arima_metrics.csv',index = False)
metrics_df
```

It seems that our considerations were correct, the classic ARIMA model with optimized parameters was able to present a more precise forecast than the Exponential Smoothing model, while still keeped aprox. 0 bias! 

Also, looking at the graphics, we see that the delay efect, heavily featured in the Exponential Smoothing method, although still present here, is much smoother! In other words, we were able to minimize the impact of outliers.
